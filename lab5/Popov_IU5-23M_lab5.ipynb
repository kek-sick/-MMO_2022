```python
import nltk
from nltk import tokenize
```

```python
text_data = None 
with open('15660.txt', 'r') as f:
    text_data = f.read()
text_data
```

```python
#Токенизация по словам
nltk_tk_1 = nltk.WordPunctTokenizer()
nltk_tk_1.tokenize(text_data)
```

```python
#Токенизация по предложениям
nltk.download('punkt')
nltk_tk_sents = nltk.tokenize.sent_tokenize(text_data)
print(len(nltk_tk_sents))
nltk_tk_sents
```

```python
#используем другую библиотеку для частиречной разметки
import spacy
#from spacy.lang.en import English
nlp = spacy.load('en_core_web_sm')
sp_tokens = nlp(text_data)
sp_tokens
```

```python
for i in range(10):
    print(sp_tokens[i])
```

```python
#Частеречная разметка
for i in range(30):
    print('{} - {} - {}'.format(sp_tokens[i].text, sp_tokens[i].pos_, sp_tokens[i].dep_))
```

```python
#Лемматизация
for i in range(30):
    print(sp_tokens[i], sp_tokens[i].lemma, sp_tokens[i].lemma_)
```

```python
#Выделение (распознавание) именованных сущностей
from spacy import displacy
displacy.render(sp_tokens, style='ent', jupyter=True)
```

```python
#Разбор предложения
sp_sentence1 = sp_tokens[:6]
sp_sentence1
```

```python
displacy.render(sp_sentence1, style='dep', jupyter=True)
```

```python
#Способ 1: TfidfVectorizer
import numpy as np
import pandas as pd
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
text_data_list = None 
with open('15660.txt', 'r') as f:
    text_data_list = f.readlines()
text_data_list
```

```python
tfidfv = TfidfVectorizer(ngram_range=(1,3))
tfidf_ngram_features = tfidfv.fit_transform(text_data_list)
tfidf_ngram_features
```

```python
# Размер нулевой строки
len(tfidf_ngram_features.todense()[0].getA1())
```

```python
# Непустые значения нулевой строки
[i for i in tfidf_ngram_features.todense()[0].getA1() if i>0]
```
